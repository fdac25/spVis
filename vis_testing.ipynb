```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
json_path = "Spotify_Extended_Streaming_History/Streaming_History_Audio_2021_9.json"
df = pd.read_json(json_path)

print(df.head())

artist_count = df["master_metadata_album_artist_name"].value_counts()
#print(artist_count)
album_count = df["master_metadata_album_album_name"].value_counts()
#print(album_count)
```

```python
json_path_kale = "Spotify_Extended_Streaming_History/Streaming_History_Audio_*.json"
dfs = []

for file in glob.glob(json_path_kale):
  data = pd.read_json(file)
  dfs.append(data)

combined_df = pd.concat(dfs, ignore_index=True)


print(combined_df.head())
num_rows = combined_df.shape[0]
print(num_rows)

'''artist_count = combined_df["master_metadata_album_artist_name"].value_counts()
print(artist_count)
album_count = combined_df["master_metadata_album_album_name"].value_counts()
print(album_count)'''
```

```python
num_rows = combined_df.shape[0]
print(num_rows)

artist_count_2 = combined_df['master_metadata_album_artist_name'].value_counts()
print(artist_count_2.head())
artist_count_top_5 = artist_count_2.head()

plt.bar(artist_count_top_5.index, artist_count_top_5.values, color=['red', 'blue', 'orange', 'purple', 'green'])
plt.xlabel('Artist')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

album_count_2 = combined_df['master_metadata_album_album_name'].value_counts()
album_count_top_5 = album_count_2.head()

plt.bar(album_count_top_5.index, album_count_top_5.values, color=['red', 'blue', 'orange', 'purple', 'green'])
plt.xlabel('Album')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
